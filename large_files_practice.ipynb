#5.Storing metadata

```python
import os
file_metadata = {}

for root, dictionaries, files in os.walk('/Users/amayatibana/Documents'):
    for file in files:
        full_path = os.path.join(root, file)
        size = os.path.getsize(full_path)
        file_metadata[full_path]= size
len(file_metadata)
```

```python
#6.Sorting and reporting

sorted_files = sorted(
    file_metadata.items(),
    key=lambda x: x[1],
    reverse = True
)

for path,size in sorted_files[:5]:
    print(size,path)
```

```python
items_shown = 0
for path, size in sorted(file_metadata.items(), key=lambda x:x[1], reverse=True):
    if items_shown > 10:
        break
    print(f"Size: {size}")
    print(f"Path: {path}")
    items_shown +=1
```
